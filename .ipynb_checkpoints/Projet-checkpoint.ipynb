### Importation de la base de données, première visualisation et explications

```python
df = pd.read_csv('csic_database.csv')
#print(df.head())

#print(df.columns.values[0])
df.columns.values[0] = 'Class'
#print(df.columns.values[0])

print(df.columns.values)
print("\n")
print('df =\n', df)
print('\n df_Info =', df.info())
```

Class : Variable catégorielle définissant le type de la requête web (normal / anormal)

User agent : Dans le cadre du Web, lorsqu'un internaute visite une page, une chaîne de caractères est généralement envoyée au serveur pour identifier l'agent utilisateur. Elle est incluse dans la requête HTTP par l'en-tête « User-Agent » et elle donne des informations comme : le nom de l'application, la version, le système d'exploitation, la langue, etc.

Method : Le type de la méthode utilisée pour la requête. Le protocole HTTP définit un certain nombre de méthodes dans les plus communes sont :
* GET : La méthode GET demande une représentation de la ressource spécifiée. Les requêtes GET sont utilisées afin de récupérer des données.
* POST : La méthode POST est utilisée pour envoyer une entité vers la ressource indiquée. Cela entraîne généralement un changement d'état sur le serveur.
* PUT : La méthode PUT remplace toutes les représentations actuelles de la ressource visée par le contenu de la requête.
* DELETE : La méthode DELETE supprime la ressource indiquée.

Pragma no-cache : Indiquer à la page web de ne pas la garder en cache. 

Cache-control no-cache : Même principe que Pragma. Cependant l'en-tête Pragma est devenue obsolète (notamment car la mise en cache n'était pas tout le temps honorée). L'en-tête Cache-control offre plus d'options sur la mise en cache. 

Accept : Définit les types de média que le client peut accepter du serveur . Par exemple, Accept: application/json, text/html indique que le client préfère les réponses JSON ou HTML.

Accept-Encoding : Permet de définir quel sera l'encodage du contenu.

Accept-Charset : Indique le jeu de caractères que le client est capable de comprendre

### Répartition entre requêtes 'normal' et 'anormal'

```python
Normal = df[df.columns[0]].value_counts().get('Normal', 0)
Anormal = df[df.columns[0]].value_counts().get('Anomalous', 0)
#print(f"Normal = {Normal}") 
#print(f"Anormal = {Anormal}")

fig, ax = plt.subplots()

Type = ['Normal', 'Anormal']
counts = [Normal, Anormal]
bar_labels = ['blue', 'red']
bar_colors = ['tab:blue', 'tab:red']

ax.bar(Type, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Nombre de requêtes')
ax.set_title('Répartition des types de requêtes')
ax.legend(title='Type')

plt.show()
```
